```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
apple_store = pd.read_csv(r"C:\Users\Lenovo\Downloads\AppleStore.csv")
```

```python
print('Lets look at the head:')
apple_store.head()
```

```python
#Top 15 most rated apps and the Top 20 best rated apps
Top15MostRated = apple_store.sort_values(by = ['rating_count_tot']).tail(15)
Top20BestRated = apple_store.sort_values(by = ['user_rating']).tail(20)
fig, axes = plt.subplots(1 , 2 , figsize = (15 , 15))
axes[0].barh(Top20BestRated['track_name'],Top20BestRated['user_rating'] , color = 'green')
axes[0].set_title("Top 20 Best Rated Apps")
axes[0].set_xlabel("App Rating")
axes[0].set_ylabel("Track Names")
axes[1].barh(Top15MostRated['track_name'] , Top15MostRated['rating_count_tot'])
axes[1].set_title("Top 15 Most Rated Apps")
axes[1].set_xlabel("Times Rated")
axes[1].set_ylabel("Track Name")
fig.tight_layout()
plt.savefig("TopRatedApps.png",bbox_inches="tight")

```
